# Excel Parser Smoke Test
Use this notebook to exercise the Excel parser against a chosen workbook.
1. Set `excel_path` below to your `.xlsx` file.
2. Run the cells to load the workbook and inspect parsed fragments.

```python
from pathlib import Path
from typing import List

from excel_parse_og import ExcelParser

excel_path = Path('sample.xlsx')  # <- change to your workbook path
excel_path
```

```python
parser = ExcelParser()
docs: List[dict] = parser.parse(excel_path, str(excel_path))
len(docs)
```

```python
def summarize_documents(documents: List[dict], max_rows: int = 5):
    summary = []
    for doc in documents[:max_rows]:
        meta = doc.get('metadata', {})
        summary.append({
            'type': meta.get('type'),
            'sheet': meta.get('sheet_name'),
            'rows': meta.get('num_rows'),
            'cols': meta.get('num_columns'),
            'table_name': meta.get('table_name'),
            'extra_keys': sorted(set(meta.keys()) - {'type', 'sheet_name', 'num_rows', 'num_columns', 'table_name'})
        })
    return summary

summarized = summarize_documents(docs, max_rows=10)
summarized
```
